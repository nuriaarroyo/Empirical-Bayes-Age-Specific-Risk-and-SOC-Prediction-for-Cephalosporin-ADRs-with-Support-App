```python
# from merge2, keep the values in REPORT_ID that are in report_unique.parquet
import polars as pl
path = "data/processed/Canada Vigilance Adverse Reaction Online Database/"
merge2 = pl.read_parquet(path + "merge2.parquet")
# not join, but select only those REPORT_ID in report
report = pl.read_parquet(path + "report_unique.parquet")
report_ids = report.select("REPORT_ID").unique()
```

```python
pair_ohe_full = pl.read_parquet(path + "pair_ohe_full.parquet")
```

```python
report_ids
```

```python
merge4 = merge2.filter(pl.col("REPORT_ID").is_in(report_ids["REPORT_ID"]))
```

```python
merge4
```

```python
# most common ACTIVE_INGREDIENT_NAME IT IS POLARS! IT DOESN HAVE THE GROUPBY FUNCTION
# most common ACTIVE_INGREDIENT_NAME - using Polars group_by
result = merge4.group_by('ACTIVE_INGREDIENT_NAME').agg(pl.count().alias('count')).sort('count', descending=True).head(30)

# plot the result an horizontal bar chart from up to down
import matplotlib.pyplot as plt
plt.barh(result['ACTIVE_INGREDIENT_NAME'].to_list(), result['count'].to_list())
plt.xlabel("Frequency")
plt.title("Top 30 Active Ingredient Names")
plt.gca().invert_yaxis()

plt.savefig('visualizations/canada/top30_active_ingredient_names.pdf')
plt.show()
```

```python
# save unique values in ACTIVE_INGREDIENT_NAME
cefs = merge4.select("ACTIVE_INGREDIENT_NAME").unique().sort("ACTIVE_INGREDIENT_NAME")
cefs.write_csv(path + "cefs4.csv")
```
