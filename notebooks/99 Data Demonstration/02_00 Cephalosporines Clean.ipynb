# Canada Vigilance Adverse Reaction Online Database

https://www.canada.ca/en/health-canada/services/drugs-health-products/medeffect-canada/adverse-reaction-database/canada-vigilance-adverse-reaction-online-database-data-structures.html

Son las tablas que creo que hay que usar

```python
import polars as pl

OUTPUT = "data/raw/Canada Vigilance Adverse Reaction Online Database/"
```

## Report Drug

```python
report_drug_filtered = pl.read_parquet(OUTPUT + "report_drug.parquet")
report_drug_filtered.write_parquet("data/cephalosporines_clean/report_drug.parquet")
report_drug_filtered
```

```python
ids = report_drug_filtered.select(pl.col("REPORT_ID")).unique()
ids = ids.sort("REPORT_ID")
ids
```

## Drug Product Ingredients

```python
dpi = pl.read_parquet(OUTPUT + "drug_product_ingredients.parquet")
dpi.write_parquet("data/cephalosporines_clean/drug_product_ingredients.parquet")
dpi
```

## Reactions

```python
reactions = pl.read_parquet(OUTPUT + "reactions.parquet")
reactions = reactions.join(ids, on="REPORT_ID", how="inner")
reactions = reactions.sort("REPORT_ID")
reactions.write_parquet("data/cephalosporines_clean/reactions.parquet")
reactions
```

## Report Drug Indications

```python
rdi = pl.read_parquet(OUTPUT + "report_drug_indication.parquet")
rdi.write_parquet("data/cephalosporines_clean/report_drug_indication.parquet")
rdi
```

## Reports Cleaned

```python
reports = pl.read_parquet("data/processed2/reports_cleaned.parquet")
reports = reports.join(ids, on="REPORT_ID", how="inner")
reports = reports.sort("REPORT_ID")
reports.write_parquet("data/cephalosporines_clean/reports_short.parquet")
reports
```

# Reports Raw

```python
reports = pl.read_parquet(OUTPUT + "reports.parquet")
reports = reports.join(ids, on="REPORT_ID", how="inner")
reports = reports.sort("REPORT_ID")
reports.write_parquet("data/cephalosporines_clean/reports_raw.parquet")
reports
```
