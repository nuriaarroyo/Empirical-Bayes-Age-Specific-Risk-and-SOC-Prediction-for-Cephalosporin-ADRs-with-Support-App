```python
# load the models
import joblib
import pandas as pd
```

```python
# load the models
models = joblib.load("interface/catboost.joblib")

# load feature names and soc names
feature_names = pd.read_csv("interface/feature_names.csv").squeeze().tolist()
soc_names = pd.read_csv("interface/soc_columns.csv").squeeze().tolist()

# get predictions for a new report
def predict_soc(report):
    report_features = report[feature_names]
    soc_probs = {}
    for soc, model in models.items():
        prob = model.predict_proba(report_features)[:, 1][0]
        soc_probs[soc] = prob
    return soc_probs    


# create an example with all features set to 0
example = pd.DataFrame(0, index=[0], columns=feature_names)

# Set specific values
example.loc[0, 'AGE_Y'] = 28
example.loc[0, 'GENDER_CODE'] = 1
example.loc[0, 'cefalexin'] = 1
example.loc[0, 'WEIGHT_KG'] = 89
example.loc[0, 'HEIGHT_CM'] = 175

# view the example
example


soc_probs = predict_soc(example)
soc_probs
```
